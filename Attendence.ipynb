```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1=df.drop(columns=['Name','Normal Working Hours',' OT Hours ','Late Coming Hours',' Early Going Hours '], axis=1)
corr=df1.corr()
sns.heatmap(corr, annot=True, cmap="coolwarm",linewidths=0.5)
plt.title("Correlation Matrix of attendence Dataset")
plt.show()
```
